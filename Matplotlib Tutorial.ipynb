```python
a = np.array([[1,2,4,5],
             [5,6,7,8]])
plt.plot(a[0], a[1] , 'go')
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')

plt.plot(a[0], a[1] , 'g--')
```

```python
t = np.linspace(0, 5 , 20)

plt.plot(t, t ** 2 ,color = 'coral', linestyle = '--')
plt.plot(t, 4 * t ,color = 'indigo', linestyle = '-')
```

```python
plt.subplot(1, 3, 1)
plt.plot(t, t  ,color = 'coral', linestyle = '--')

plt.subplot(1, 3, 2)
plt.plot(t, t ** 2 ,color = 'blue', linestyle = '-')

plt.subplot(1, 3, 3)
plt.plot(t, t ** 3 ,'mo')
```

```python
grades = {'John' : 70,
          'Johny' : 82,
          'Merry' : 92,
          'Myra' : 76,
          'Alexa' : 96
}

plt.bar(range(len(grades)), list(grades.values()), color = 'darkturquoise')
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel("Students")
plt.ylabel("Grades")
```
